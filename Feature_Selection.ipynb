**Univariate Selection**

```python
# Univariate selection
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectPercentile
iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=0, test_size=0.5)
select = SelectPercentile(percentile=50).fit(X_train, y_train)
X_train_select=select.transform(X_train)

print(X_train.shape)
print(X_train_select.shape)
```

 **Model Based Selection**

```python
#Model Based Selection
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=0, test_size=0.5)
select = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42),threshold="median").fit(X_train, y_train)
X_train_select=select.transform(X_train)
X_test_select=select.transform(X_test)
print(X_train.shape)
print(X_train_select.shape)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(max_iter=100000).fit(X_train,y_train)
# accuracy before selection
print("LR training Accuracy is : ",LR.score(X_train, y_train))
print("LR testing Accuracy is : ",LR.score(X_test, y_test))
# accuracy after selection
LR1 = LogisticRegression(max_iter=100000).fit(X_train_select,y_train)
print("LR1 training Accuracy is : ",LR1.score(X_train_select, y_train))
print("LR1 testing Accuracy is : ",LR1.score(X_test_select, y_test))
```

# **Iterative Feature Selection**
Recursive Feature Selection

```python
# Iterative Feature Selection
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0, test_size=0.5)
select = RFE(RandomForestClassifier(n_estimators=100, random_state=42),n_features_to_select=15).fit(X_train, y_train)
X_train_select=select.transform(X_train)
X_test_select=select.transform(X_test)
print(X_train.shape)
print(X_train_select.shape)
LR = LogisticRegression(max_iter=100000).fit(X_train,y_train)
# accuracy before selection
print("LR training Accuracy is : ",LR.score(X_train, y_train))
print("LR testing Accuracy is : ",LR.score(X_test, y_test))
# accuracy after selection
LR1 = LogisticRegression(max_iter=100000).fit(X_train_select,y_train)
print("LR1 training Accuracy is : ",LR1.score(X_train_select, y_train))
print("LR1 testing Accuracy is : ",LR1.score(X_test_select, y_test))
# Train with select
print("LR training Accuracy is : ",select.score(X_train, y_train))
print("LR testing Accuracy is : ",select.score(X_test, y_test))
```

# **Principle Component Analysis (PCA)**

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
cancer = load_breast_cancer()

# Apply PCA to reduce dimensions to 2
X_pca = PCA(n_components=2).fit_transform(cancer.data)

# Split the PCA transformed data
X_train, X_test, y_train, y_test = train_test_split(X_pca, cancer.target, random_state=0, test_size=0.5)

# Train a Logistic Regression model on the PCA-transformed data
LR = LogisticRegression(max_iter=100000).fit(X_train, y_train)

# Accuracy after PCA
print("LR training Accuracy (after PCA) is : ", LR.score(X_train, y_train))
print("LR testing Accuracy (after PCA) is : ", LR.score(X_test, y_test))
```
