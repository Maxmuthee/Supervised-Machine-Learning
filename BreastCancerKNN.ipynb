```python
import mglearn
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 66) # split the rows(data) for the training and the testing

#creating the model
KNN = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)#create a model with training data
print(KNN.score(X_test, y_test))
#print(f"{(KNN.score(X_test, y_test) * 100):.2f}%")
print(KNN.predict(X_test))
```
