```python
import pandas as pd
income = pd.read_excel("DataTransformation.xlsx")
display(income.head())
```

```python
e_income = pd.get_dummies(income,dtype=int)
display(e_income.head())
```

```python
education_num = pd.get_dummies(income['education-num'], dtype=int)
display(education_num.head())
```

```python
#deserialization
import numpy as np
bins = np.linspace(max(e_income["age"]), min(e_income["age"]),5)#no of groupings
bins
```

```python
e_income["b_age"]=np.digitize(e_income["age"],bins=bins)
print(e_income)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X=e_income.drop(columns=["income_ <=50K","income_ >50K"])
y=e_income["income_ <=50K"]
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state = 0)# you can ignore the random state unless asked otherwise
LR = LogisticRegression(max_iter=1000).fit(X_train, y_train)#.fit() is for training
print("Training Accuracy: ", LR.score(X_train, y_train))
print("Testing Accuracy: ", LR.score(X_test, y_test))
```
