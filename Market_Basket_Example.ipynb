```python
# Load Dataset
import pandas as pd
data = pd.read_csv("Market_Basket_Optimisation.csv")

# printing the shape of the dataset
data
```

```python
# Gather All Items of Each Transactions into Numpy Array
import numpy as np
transactions = []

for i in range(data.shape[0]):
    transaction = []
    for j in range(data.shape[1]):
        item = str(data.values[i, j])
        if item != 'nan':
            transaction.append(item)
    transactions.append(transaction)
```

```python
# Flatten list of lists into one list of items
transactions_flat = [item for sublist in transactions for item in sublist]

# Create DataFrame
df = pd.DataFrame(transactions_flat, columns=["items"])

# Add incident count
df["incident_count"] = 1

# Drop NaN values
df = df[df["items"] != "nan"]

# Frequency table
df_table = df.groupby("items").sum().sort_values("incident_count", ascending=False).reset_index()

# Visualization
df_table.head(10).style.background_gradient(cmap='Greens')
```

```python
# importing the required module
from mlxtend.preprocessing import TransactionEncoder
# initializing the transactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
dataset = pd.DataFrame(te_ary, columns=te.columns_)
# dataset after encoded
dataset
```

```python
# 121 items are too many and some occur a few times only
# so we select the most common

# select top 50 items
first50 = df_table["items"].head(50).values

# Create dataset with Top50
dataset = dataset.reindex(columns = first50)
dataset
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
# importing the required module
from mlxtend.frequent_patterns import apriori, association_rules


# Extracting the the rules (most frequest items)
frequent_itemsets = apriori(dataset, min_support=0.01, use_colnames=True)

# We add the length column simplify filtering filtering.
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
# printing the frequent itemset
frequent_itemsets
```

```python
# printing with more than two items and support above 0.05
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.05) ]
```

```python
# Searching for specific rule
# in sets the order does not matter
frequent_itemsets[ (frequent_itemsets['itemsets'] == {'eggs', 'mineral water'}) ]
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

#convert rules to lists
rules = rules.sort_values(['lift'], ascending =[False])

#From Frozenset to string
rules["antecedents"] = rules["antecedents"].apply(lambda x: list(x)[0]).astype("unicode")
rules["consequents"] = rules["consequents"].apply(lambda x: list(x)[0]).astype("unicode")
#Table with most relevant rules
rules.sort_values('lift', ascending=False).nlargest(10, 'lift')
```
