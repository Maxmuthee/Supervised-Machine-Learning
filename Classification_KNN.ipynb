```python
from sklearn.model_selection import train_test_split # Helps us split the training data and the testing data in terms of columns i.e. the features and the classes .
from sklearn.neighbors import KNeighborsClassifier #Importing of the library and the model (K-NN)
import mglearn
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0) # split the rows(data) for the training and the testing

#creating the model
KNN = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)#create a model with training data
print(KNN.score(X_test, y_test)*100)
#print(f"{(KNN.score(X_test, y_test) * 100):.2f}%")
print(KNN.predict(X_test))
```

```python
#using real datasets
from sklearn.datasets import load_iris
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state = 66) # split the rows(data) for the training and the testing

#creating the model
KNN = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)#create a model with training data
print(KNN.score(X_test, y_test))
#print(f"{(KNN.score(X_test, y_test) * 100):.2f}%")
print(KNN.predict(X_test))
```
