```python
#Stacking
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target, random_state = 66)
LR = LogisticRegression(max_iter=100000).fit(X_train,y_train)
KNN =KNeighborsClassifier(n_neighbors = 3).fit(X_train,y_train)
DT = DecisionTreeClassifier().fit(X_train, y_train)
ENSEMBLE = StackingClassifier(estimators = [('LR',LR),('KNN', KNN),('DT',DT)], final_estimator = LogisticRegression()).fit(X_train,y_train)
print("LR training Accuracy is : ",LR.score(X_train, y_train))
print("LR testing Accuracy is : ",LR.score(X_test, y_test))
print("DT training Accuracy is : ",DT.score(X_train, y_train))
print("DT testing Accuracy is : ",DT.score(X_test, y_test))
print("KNN training Accuracy is : ",KNN.score(X_train, y_train))
print("KNN testing Accuracy is : ",KNN.score(X_test, y_test))
print("ENSEMBLE training Accuracy is : ",ENSEMBLE.score(X_train, y_train))
print("ENSEMBLE testing Accuracy is : ",ENSEMBLE.score(X_test, y_test))
```

```python
#Majority Voting
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target, random_state = 66)
LR = LogisticRegression(max_iter=100000).fit(X_train,y_train)
KNN =KNeighborsClassifier(n_neighbors = 3).fit(X_train,y_train)
DT = DecisionTreeClassifier().fit(X_train, y_train)
ENSEMBLE = VotingClassifier(estimators = [('LR',LR),('KNN', KNN),('DT',DT)], voting = 'hard').fit(X_train,y_train)
print("LR training Accuracy is : ",LR.score(X_train, y_train))
print("LR testing Accuracy is : ",LR.score(X_test, y_test))
print("DT training Accuracy is : ",DT.score(X_train, y_train))
print("DT testing Accuracy is : ",DT.score(X_test, y_test))
print("KNN training Accuracy is : ",KNN.score(X_train, y_train))
print("KNN testing Accuracy is : ",KNN.score(X_test, y_test))
print("ENSEMBLE training Accuracy is : ",ENSEMBLE.score(X_train, y_train))
print("ENSEMBLE testing Accuracy is : ",ENSEMBLE.score(X_test, y_test))
```

```python
#Bagging
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target, random_state = 66)
DT = DecisionTreeClassifier().fit(X_train, y_train)
ENSEMBLE = BaggingClassifier(estimator = DT, n_estimators = 10 ).fit(X_train,y_train)
print("DT training Accuracy is : ",DT.score(X_train, y_train))
print("DT testing Accuracy is : ",DT.score(X_test, y_test))
print("ENSEMBLE training Accuracy is : ",ENSEMBLE.score(X_train, y_train))
print("ENSEMBLE testing Accuracy is : ",ENSEMBLE.score(X_test, y_test))
```

```python
#Bagging - Random Forest
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target, random_state = 66)
ENSEMBLE = RandomForestClassifier(n_estimators = 300, max_features = 10, max_depth = 4 ).fit(X_train,y_train)
print("ENSEMBLE training Accuracy is : ",ENSEMBLE.score(X_train, y_train))
print("ENSEMBLE testing Accuracy is : ",ENSEMBLE.score(X_test, y_test))
```

```python
#Boosting
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target, random_state = 66)
DT = DecisionTreeClassifier().fit(X_train, y_train)
ENSEMBLE = AdaBoostClassifier(estimator = DT, n_estimators = 10 ).fit(X_train,y_train)
print("DT training Accuracy is : ",DT.score(X_train, y_train))
print("DT testing Accuracy is : ",DT.score(X_test, y_test))
print("ENSEMBLE training Accuracy is : ",ENSEMBLE.score(X_train, y_train))
print("ENSEMBLE testing Accuracy is : ",ENSEMBLE.score(X_test, y_test))
```
